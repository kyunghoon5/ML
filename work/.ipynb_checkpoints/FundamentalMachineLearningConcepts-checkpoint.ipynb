### Separation of numerical and categorical variables

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
target_variable = 'mpg'
# ignore model_year as it can be more like categorical variables
numerical_features = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']
categorical_features = ['origin', 'name']

numerical_df = df[[target_variable] + numerical_features]
numerical_df.head()
```

***Calculate Correlation***

```python
numerical_df.corr()
```

```python
categorical_df = df[categorical_features]
categorical_df.head()
```

### Features Scaling

```python
df.describe()
```

**Note that all the features are of different scale, weight is in thousands and horsepower is around 100**

**Feature scaling is to "standardize" all features to be around the same magnitude**

```python
from   import StandardScaler, MinMaxScaler
```

```python
# drop the target variable, keep only the numerical features
num_feature_df = numerical_df.drop('mpg', axis=1)
num_feature_df.head()
```

```python
scaler = StandardScaler()
normalized_df = scaler.fit_transform(num_feature_df)
normalized_df = pd.DataFrame(normalized_df, index = df.index, columns = ["cylinders","displacement","horsepower","weight", "acceleration"] )
normalized_df.head()
```

```python
normalized_df.tail()
```

```python
normalized_df.describe()
```

### Now all features has mean of 0 and standard deviation of 1

```python
minMaxScaler = MinMaxScaler()
minMax_df = minMaxScaler.fit_transform(num_feature_df)
minMax_df = pd.DataFrame(minMax_df, columns = ["cylinders","displacement","horsepower","weight", "acceleration"] )
minMax_df.head()
```

```python
minMax_df.describe()
```

***Now you can use the normalized_df as your independent variables***

## How to include Categorical Variables using one-hot encoding

https://www.kaggle.com/camnugent/california-housing-prices

```python
df = pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
# ignore longitude and latitude
numerical_features = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households','median_income','median_house_value']
categorical_features = ['ocean_proximity']
```

```python
numerical_df = df[numerical_features]
```

```python
numerical_df.corr()
```

```python
sns.pairplot(numerical_df)
```

***Median income has the highest correlation with median_house_value, so we will use it as our independent variable***

```python
sns.scatterplot(x='median_income', y='median_house_value', data=numerical_df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn import metrics


X = numerical_df['median_income'].values.reshape(-1, 1)
Y = numerical_df['median_house_value'].values.reshape(-1, 1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print(model.intercept_)
print(model.coef_)
print('R-squared:', metrics.r2_score(Y_test, Y_pred))
```

```python
plt.scatter(X_test, Y_test,  color='gray')
plt.scatter(X_test, Y_pred, color='red', linewidth=2)
plt.show()
```

***We can add more numerical variables to fit the house price***

```python
X = numerical_df[['median_income', 'total_rooms']].values.reshape(-1, 2)
Y = numerical_df['median_house_value'].values.reshape(-1, 1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print(model.intercept_)
print(model.coef_)
print('R-squared:', metrics.r2_score(Y_test, Y_pred))
```

***So it doesn't really help. Can we use the categorical variable?***

```python
df['ocean_proximity'].unique()
```

```python
categorical_df = df[categorical_features]
categorical_df.head()
```

```python
categorical_df.tail()
```

```python
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder()
housing_cat_1hot = cat_encoder.fit_transform(categorical_df)
housing_cat_1hot
```

```python
housing_cat_1hot.toarray()
```

```python
cat_encoder.categories_
```

```python
onehot_df = pd.DataFrame(housing_cat_1hot.toarray(), columns = cat_encoder.categories_)
```

```python
onehot_df.head()
```

```python
onehot_df.tail()
```

```python
combine_df = pd.concat([numerical_df[['median_income', 'total_rooms','median_house_value']], onehot_df ], axis=1)
```

```python
combine_df.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(combine_df.drop('median_house_value',axis=1), 
                                                    combine_df['median_house_value'], 
                                                    test_size=0.30, 
                                                    random_state=101)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print(model.intercept_)
print(model.coef_)
print('R-squared:', metrics.r2_score(Y_test, Y_pred))
```

***R-squared has been increased from 45% to 58%!!!***

```python
X_train, X_test, Y_train, Y_test = train_test_split(combine_df.drop('median_house_value',axis=1), 
                                                    combine_df['median_house_value'], 
                                                    test_size=0.30, 
                                                    random_state=101)
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print(model.intercept_)
print(model.coef_)
print('R-squared:', metrics.r2_score(Y_test, Y_pred))
```

### Use ordinal encoding to handle categorical variable (NOT recommended)

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
housing_cat_encoded = ordinal_encoder.fit_transform(categorical_df)
housing_cat_encoded[:10]
```

```python
ordinal_encoder.categories_
```

```python
categorical_df
```

```python
ordinal_encode_df = pd.DataFrame(housing_cat_encoded, columns = ["ocean_ordinal"])
ordinal_encode_df.head()
```

```python
combine_df3 = pd.concat([numerical_df[['median_income', 'total_rooms','median_house_value']], ordinal_encode_df ], axis=1)
combine_df3.head()
```

***Note there are only 1 additional features from ordinal encoding***

```python
X_train, X_test, Y_train, Y_test = train_test_split(combine_df3.drop('median_house_value',axis=1), 
                                                    combine_df3['median_house_value'], 
                                                    test_size=0.30, 
                                                    random_state=101)
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print(model.intercept_)
print(model.coef_)
print('R-squared:', metrics.r2_score(Y_test, Y_pred))
```

***Note using ordinal encoding result only a slight improvement from 45% to 47%***
